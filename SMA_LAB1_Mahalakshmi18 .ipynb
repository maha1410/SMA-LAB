### 6. Print common trends using set function.

```python
India_trends_set = set([trend['name'] 
                        for trend in India_trends[0]['trends']])

Lucknow_trends_set = set([trend['name'] 
                     for trend in Lucknow_trends[0]['trends']]) 

common_trends = India_trends_set.intersection(Lucknow_trends_set)

print(common_trends)
```
